```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
total_loss = pd.read_csv('results/run_=training_.-tag-summary_total_loss.csv')
```

```python
axis = total_loss['Step']
total = total_loss['Value']
```

```python
plt.plot(axis,total)
plt.xlabel('Iterations')
plt.ylabel('Total Loss')
```

```python
content_loss = pd.read_csv('results/run_=training_.-tag-summary_content_loss.csv')
axis = content_loss['Step']
content = content_loss['Value']
```

```python
plt.plot(axis,content)
plt.xlabel('Iterations')
plt.ylabel('Content Loss')
```

```python
style_loss = pd.read_csv('results/run_=training_.-tag-summary_style_loss.csv')
axis = style_loss['Step']
style = style_loss['Value']
```

```python
plt.plot(axis,style)
plt.xlabel('Iterations')
plt.ylabel('Style Loss')
```

```python
plt.plot(axis,content, label="Content Loss")
plt.plot(axis,style, label="Style Loss")
plt.plot(axis,total, label="Total Loss")
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()
```
